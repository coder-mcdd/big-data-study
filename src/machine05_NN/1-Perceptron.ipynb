# 感知机

感知机（Perceptron）是二分类的线性分类模型其输入为实例的特征向量输出为实例的类别（取`+1`或`-1`）感知机的学习过程就是 
1. 将实例通过分离超平面将特征空间划分为两类，为求得超平面导入了基于误分类的损失函数
2. 利用梯度下降法对损失函数进行最优化

## 1. 生物学、心理学解释

心理学家唐纳德·赫布（Donald Olding Hebb）于 1949 年提出`赫布理论`，该理论能够解释学习的过程中脑中的神经元所发生的变化。赫布理论描述了突触可塑性的基本原理，即 **突触前神经元向突触后神经元的持续重复的刺激，可以导致突触传递效能的增加**：
> 当细胞 A 的轴突足以接近以激发细胞 B，并反复持续地对细胞 B 放电，一些生长过程或代谢变化将发生在某一个或这两个细胞内，以致A 作为对 B 放电的细胞中的一个，导致突触传递效能的增加。

![neuron_cell](images/neuron_cell_cn.png)


心理学家弗兰克·罗森布拉特(Frank Rosenblatt)于 1957 年提出了`感知机`，它作为简化的数学模型解释大脑神经元如何工作：它取一组二进制输入值（附近的神经元），将每个输入值乘以一个连续值权重（每个附近神经元的突触强度），并设立一个阈值，如果这些加权输入值的和超过这个阈值，就输出 1，否则输出 0，这样的工作原理就相当于神经元是否放电。


感知机并没有完全遵循赫布理论，**但通过调输入值的权重，可以有一个非常简单直观的学习方案：给定一个有输入输出实例的训练集，感知机应该「学习」一个函数：对每个例子，若感知机的输出值比实例低太多，则增加它的权重；否则若设比实例高太多，则减少它的权重。**


**感知机模仿的是生物神经系统内的神经元，它能够接受来自多个源的信号输入，然后将信号转化为便于传播的信号在进行输出(在生物体内表现为电信号)**

## 2. 感知机模型

假设输入空间(特征向量)为 $\mathbf{X} \subseteq \mathbb{R}^D$，输出空间为 $\mathbf{Y} \in \{-1, +1\}$。输入 $x \in \mathbf{X}$ 表示实例的特征向量，对应于输入空间的点；输出 $y \in \mathbf{Y}$ 表示示例的类别。由输入空间到输出空间的函数为

$$
f(x) = sign(w x + b)
$$


![sign_function](images/sign.png)




### 2.1 几何解释    
感知机模型是`线性分类模型`，感知机模型的假设空间是定义在特征空间中的所有线性分类模型，即函数集合 $\{ f | f(x)=w·x+b\}$。线性方程 $w·x+b=0$ 对应于特征空间 $\mathbb{R}^D$ 中的一个超平面 $S$，其中 $w$ 是超平面的法向量，$b$ 是超平面的截距。这个超平面把特征空间划分为两部分，位于两侧的点分别为正负两类。超平面 $S$ 称为分离超平面，如下图：

![perceptron_geometry_def](images/perceptron_geometry_def.png)



## 3. 感知机学习策略

**假设训练数据集是线性可分的，感知机学习的目标是求得一个能够将训练数据的正负实例点完全分开的分离超平面，即最终求得参数 $w, b$。这需要一个学习策略，即定义（经验）损失函数并将损失函数最小化 损失函数的一个自然的选择是误分类的点的总数，但是这样得到的损失函数不是参数 $w，b$ 的连续可导函数，不宜优化。损失函数的另一个选择是误分类点到分类面的距离之和。**

首先，对于任意一点 $x$ 到超平面的距离（[参考资料](https://www.cnblogs.com/graphics/archive/2010/07/10/1774809.html)）为

$$
\frac{1}{||w||} | w \cdot x + b |
$$

其次，对于误分类点 $(x_i,y_i)$ 来说 $-y_i(w \cdot x_i + b) > 0$

这样，假设超平面 $S$ 的总的误分类点集合为 $\mathbf{M}$，那么所有误分类点到 $S$ 的距离之和为

$$
-\frac{1}{||w||} \sum_{x_i \in \mathbf{M}} y_i (w \cdot x_i + b)
$$

不考虑 $1/||w||$，就得到了感知机学习的损失函数

$$
L = - \sum_{x_i \in \mathbf{M}} y_i (w \cdot x_i + b)
$$

### 3.1 经验风险函数

给定数据集 $\mathbf{T} = \{(x_1,y_1), (x_2, y_2), ... (x_N, y_N)\}$，其中

* $x_i \in \mathbb{R}^D$
* $y_i \in \{-1, +1\}，i=1,2...N$

感知机 $sign(w·x+b)$ 学习的损失函数定义为

$$
L(w, b) = - \sum_{x_i \in \mathbf{M}} y_i (w \cdot x_i + b)
$$

其中 $\mathbf{M}$ 为误分类点的集合，这个损失函数就是感知机学习的[经验风险函数](https://blog.csdn.net/zhzhx1204/article/details/70163099)。

显然，损失函数 $L(w,b)$ 是非负的。
* 如果没有误分类点，那么 $L(w,b)$为 0
* 误分类点数越少，$L(w,b)$ 值越小

该损失函数在误分类时是参数 $w,b$ 的线性函数，在正确分类时该损失函数是 0。因此，给定训练数据集 $\mathbf{T}$，损失函数 $L(w,b)$ 是 $w,b$ 的连续可导函数。


## 4. 感知机学习算法

最优化问题：给定数据集 $\mathbf{T} = \{(x_1,y_1), (x_2, y_2), ... (x_N, y_N)\}$，其中

* $x_i \in \mathbb{R}^D$
* $y_i \in \{-1, +1\}，i=1,2...N$

求参数 $w,b$，使其成为损失函数的解，其中 $\mathbf{M}$ 为误分类数据的集合：

$$
min_{w,b} L(w, b) =  - \sum_{x_i \in \mathbf{M}} y_i (w \cdot x_i + b)
$$

感知机学习是误分类驱动的，具体采用随机梯度下降法：
* 首先，任意选定 $w_0$、$b_0$，
* 然后用梯度下降法不断极小化目标函数
   - 极小化的过程不是一次性的把 $\mathbf{M}$ 中的所有误分类点梯度下降
   - 而是一次随机选取一个误分类点使其梯度下降。

假设误分类集合 $\mathbf{M}$ 是固定的，那么损失函数 $L(w,b)$ 的梯度为

$$
\triangledown_w L(w, b) = - \sum_{x_i \in \mathbf{M}} y_i x_i \\
\triangledown_b L(w, b) = - \sum_{x_i \in \mathbf{M}} y_i \\
$$

随机选取一个误分类点 $(x_i,y_i)$,对 $w,b$ 进行更新：

\begin{eqnarray}
w & \leftarrow & w + \eta y_i x_i \\
b & \leftarrow & b + \eta y_i
\end{eqnarray}

式中 $\eta$（0 ≤ $ \eta $ ≤ 1）是学习速率（步长）：

* 步长越大，梯度下降的速度越快，更能接近极小点。如果步长过大，有可能导致跨过极小点，导致函数发散；
* 如果步长过小，有可能会耗很长时间才能达到极小点。

> 直观解释：当一个实例点被误分类时，调整 $w,b$，使分离超平面向该误分类点的一侧移动，以减少该误分类点与超平面的距离，直至超越该点被正确分类。



### 4.1 算法


**输入：**

* $\mathbf{T}=\{(x_1,y_1),(x_2,y_2), ..., (x_N,y_N)\}$, 其中$x_i \in \mathbf{X}=\mathbb{R}^D$，
* $y_i \in \mathbf{Y} = {-1, +1}，i=1,2...N$，

* 学习速率为 η

**输出：**

* $w$, $b$; 
* 感知机模型 $f(x)=sign(w·x+b)$

**处理过程:**

1. 初始化 $w_0$,$b_0$
2. 在训练数据集中选取 $(x_i, y_i）$
3. 如果 $y_i(w * x_i+b)≤0$
           
    $w = w + η y_i x_i$
    
    $b = b + η y_i$

4. 如果所有的样本都正确分类，或者迭代次数超过设定值，则终止
5. 否则，跳转至（2）


## 5. 示例程序

生成数据：目的是生成并绘制一个二维数据集，其中包含两组明显区分的数据点，用不同的颜色表示。具体步骤和结果解释如下：

    数据生成：代码首先生成了两组数据x1和x2，每组包含10个数据点。数据点x1围绕点[2,2]生成，而x2则围绕点[8,8]生成。x1的数据点从标准正态分布生成，而x2的数据点从标准差为2的正态分布生成，并移动到[8,8]。这样做的目的是让两组数据在图形上有明显的空间分布差异。

    标签分配：每组数据都被分配了标签，x1的数据点被标记为-1，而x2的数据点被标记为1。

    数据合并与随机打乱：两组数据及其标签合并后，整个数据集被随机打乱，以消除任何潜在的排序偏差。

    数据重标记：为了适应特定的数据处理需求，代码将所有-1标签替换为2。

    数据绘制：使用matplotlib库，代码绘制了一个散点图，其中x和y坐标对应于数据点的两个特征，颜色通过label_y数组指定，颜色映射为plt.cm.Spectral。这样，两类数据点以不同颜色显示，反映出它们的类别区分。

    图像保存与显示：绘制的图像被保存为PDF文件，并在屏幕上显示。

绘制出来的图像展示了两个明显分离的点群，颜色不同表示不同类别，这种类型的数据常用于机器学习中的分类任务，特别是测试分类算法的效果。

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

# data generation
np.random.seed(314)

data_size1 = 10
"""
调用 NumPy 的 randn 函数，生成一个形状为 (data_size1, 2) 的数组。即生成一个 10 行 2 列的数组，其中的元素是从标准正态分布（均值为 0，标准差为 1 的高斯分布）中随机抽取的。每行代表一个数据点，在这个上下文中，可以理解为每个数据点有两个特征，然后通过 np.array([2,2]) 创建了一个包含两个元素的 NumPy 数组 [2, 2]，将其加到前面生成的每个数据点上
结果：
0,2.166085,2.781964
1,2.852285,1.292929
2,1.068343,2.886661
3,1.778210,2.381724
4,1.227425,2.862987
5,1.718740,1.068075
6,1.501463,2.764325
7,2.191504,1.381236
8,3.665719,3.738188
9,3.182226,3.119376
"""
x1 = np.random.randn(data_size1, 2) + np.array([2,2])
"""
通过列表推导式（list comprehension）生成一个列表，其中包含 10 个 -1
[-1 for _ in range(data_size1)] 中的 range(data_size1) 生成一个从 0 到 9 的序列而对于序列中的每个元素（在这里用 _ 表示，因为我们不需要用到这个值），都放入 -1 ,这个列表 y1 作为标签与前面的数据点 x1 相对应
即生成了一个包含 10 个二维数据点的集合 x1，这些数据点集中在点 [2,2] 周围，每个点都标记为类别 -1，存储在 y1 中
"""
y1 = [-1 for _ in range(data_size1)]

data_size2 = 10
x2 = np.random.randn(data_size2, 2)*2 + np.array([8,8])
y2 = [1 for _ in range(data_size2)]

"""
all sample data
np.concatenate((x1, x2), axis=0) 用于按行将 x1 和 x2 这两个数组合并成一个新的数组 x，方便后续的数据处理和分析，其中参数
- axis=0：这指定了合并的轴向。在 NumPy 中，axis=0 通常是指沿着行的方向，即将一个数组叠加到另一个数组的下方。如果是 axis=1，则表示沿着列方向合并，即将一个数组拼接到另一个数组的右侧。
 - x1 和 x2：这两个变量代表要合并的两个 NumPy 数组。这些数组应该具有相同的列数，但行数可以不同，因为合并是沿着行进行的
 
生成的 x 和 y 在位置上还是一一对应的
"""
x = np.concatenate((x1, x2), axis=0)
y = np.concatenate((y1, y2), axis=0)

"""
通过索引 shuffled_index，x 和 y 中的每一行都会根据这个随机生成的新顺序进行重排。这意味着每个数据点及其对应的标签都会随机移动到新的位置，但数据点和标签之间的一一对应关系得以保持。
"""
shuffled_index = np.random.permutation(data_size1 + data_size2)
x = x[shuffled_index]
y = y[shuffled_index]

"""
目的：这行代码的目的是将特征集 x 和标签集 y 合并为一个单独的数据集，以便在模型训练时同时处理特征和标签。
操作解释：

    y[:, np.newaxis]：这部分代码将 y 从一个一维数组转换成一个二维数组（列向量）。np.newaxis 用于增加一个新的轴，这使得 y 变成一个形状为 (n, 1) 的数组，其中 n 是 y 中元素的数量。
    np.concatenate((x, y[:, np.newaxis]), axis=1)：这个函数将 x 和新形状的 y 沿着第二个轴（列）进行合并。这意味着每个特征向量后面都会附加其对应的标签，从而在同一个数组中并排存储特征和标签。
"""
train_data = np.concatenate((x, y[:, np.newaxis]), axis=1)

"""
目的：这部分代码用于修改标签数据，通常是为了适应特定的算法或标准。
操作解释：

    label_y = y.copy()：首先，创建 y 的一个副本，存储在 label_y 中。这样做是为了保留原始的 y 数据，同时在 label_y 上进行修改而不影响 y。
    label_y[label_y==-1] = 2：这行代码查找 label_y 中所有等于 -1 的元素，并将它们替换为 2。这种类型的操作常见于需要将特定标签编码为模型可以更好处理的形式。例如，有些模型可能预期非负整数标签或从零开始的连续整数。
"""
label_y = y.copy()
label_y[label_y==-1] = 2;
print(label_y)
# plot data
plt.scatter(train_data[:,0], train_data[:,1], marker='.',  s = 300,
            c=label_y, cmap=plt.cm.Spectral)
plt.title("Data")
plt.savefig("./temp/fig-res_perceptron_sample_data.pdf")
plt.show()
```

学习模型：




```python
import random
import numpy as np


def sign(v):
    """符号函数"""
    if v >= 0:  return 1
    else:       return -1

def perceptron_train(train_data, eta=0.01, n_iter=100):
    """对感知机模型进行训练"""
    weight = [0, 0]      # 权重
    bias = 0             # 偏置量
    learning_rate = eta  # 学习速率

    train_num = n_iter  # 迭代次数

    for i in range(train_num):
        # select one data
        ti = np.random.randint(len(train_data))
        (x1, x2, y) = train_data[ti]

        y_pred = sign(weight[0] * x1 + weight[1] * x2 + bias)

        if y * y_pred <= 0:  # 判断误分类点
            weight[0] = weight[0] + learning_rate * y * x1  # 更新权重
            weight[1] = weight[1] + learning_rate * y * x2
            bias      = bias      + learning_rate * y       # 更新偏置量
            print("update weight/bias: ", weight[0], weight[1], bias)

    return weight, bias

def perceptron_pred(data, w, b):
    """输入数据，模型，对数据进行分类"""
    y_pred = []
    for d in data:
        x1, x2, y = d
        yi = sign(w[0]*x1 + w[1]*x2 + b)
        y_pred.append(yi)

    return np.array(y_pred, dtype=float)


# 训练感知机
w, b = perceptron_train(train_data, n_iter=500)
print("w = ", w)
print("b = ", b)

# 预测 
y_pred = perceptron_pred(train_data, w, b)

# 计算分类精度
c = y_pred == y
cn = np.sum(c == True)
acc = cn / len(y_pred)
print()

print("\n")
print("ground_truth: ", train_data[:, 2])
print("predicted:    ", y_pred)
print("accuracy:     ", acc)
```

```python
label_y = y_pred.copy()
label_y[label_y==-1] = 2;

# plot data
plt.scatter(train_data[:,0], train_data[:,1], marker='.',  s = 300,
            c=label_y, cmap=plt.cm.Spectral)
plt.title("Data")
plt.savefig("./temp/fig-res_perceptron_res_data.pdf")
plt.show()
```
